```python
import git
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')

# Function to extract keywords
def extract_keywords(tokens):
    stop_words = set(stopwords.words('english'))
    keywords = [word for word in tokens if word not in stop_words and word.isalpha()]
    return keywords

repo_path = r'C:\Users\ml\PycharmProjects\LLM-Dec-Hackathon\tests\test_repo'
repo = git.Repo(repo_path)

for commit in repo.iter_commits():
    commit_data = {
        'commit_id': commit.hexsha,
        'author': commit.author.name,
        'date': commit.committed_datetime,
        'message': commit.message
    }
    tokens = word_tokenize(commit_data['message'])
    keywords = extract_keywords(tokens)

    print(commit_data)
    print("Keywords:", keywords)
    print("---")
    files_changed = {'added': [], 'modified': [], 'deleted': []}
    for file in commit.stats.files:
        if commit.stats.files[file]['insertions'] > 0 and commit.stats.files[file]['deletions'] == 0:
            files_changed['added'].append(file)
        elif commit.stats.files[file]['deletions'] > 0 and commit.stats.files[file]['insertions'] == 0:
            files_changed['deleted'].append(file)
        else:
            files_changed['modified'].append(file)

    print("Files Changed:", files_changed)
    print("---")

```

```python
import ast

class CodeAnalyzer(ast.NodeVisitor):
    def __init__(self):
        self.classes = []
        self.methods = []

    def visit_ClassDef(self, node):
        self.classes.append(node.name)
        self.generic_visit(node)

    def visit_FunctionDef(self, node):
        self.methods.append(node.name)
        self.generic_visit(node)

def analyze_python_code(file_path):
    analyzer = CodeAnalyzer()
    with open(file_path, 'r') as file:
        content = file.read()
        tree = ast.parse(content)
        analyzer.visit(tree)
    return analyzer.classes, analyzer.methods

# Example usage
classes, methods = analyze_python_code(repo_path + r'\apps\finance\apps.py')
print("Classes:", classes)
print("Methods:", methods)
```
